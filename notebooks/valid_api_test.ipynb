```python
import sys
import os
import requests
import json

# 1. Add the 'src' directory to Python's path so we can import config
# (This allows us to import from the src folder even though we are in notebooks)
sys.path.append(os.path.abspath(os.path.join('..')))

from src.config import ODDS_API_KEY, SPORT, REGIONS, MARKETS, ODDS_FORMAT

# 2. Define the Endpoint
url = f'https://api.the-odds-api.com/v4/sports/{SPORT}/odds'

# 3. Set Parameters
params = {
    'apiKey': ODDS_API_KEY,
    'regions': REGIONS,
    'markets': MARKETS,
    'oddsFormat': ODDS_FORMAT,
}

# 4. Fetch Data
response = requests.get(url, params=params)

# 5. Check Response
if response.status_code == 200:
    print("Success! API is working.")
    data = response.json()
    # Print the first game nicely formatted
    if data:
        print(f"Found {len(data)} upcoming games.")
        print(json.dumps(data[0], indent=2))
    else:
        print("API worked, but no games are scheduled right now.")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
